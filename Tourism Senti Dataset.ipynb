# Tourism Senti Dataset

```python
import pandas as pd

# Read data
review_df = pd.read_excel("Tourism Senti Dataset.xlsx")
review_df.head()
```

### NLTK (NLP - Natural Language Processing)
an acronym for Natural Language Toolkit

We use Vader, which is a part of the NLTK module designed for sentiment analysis

For each text, Vader returns 4 values:

- a neutrality score
- a positivity score
- a negativity score
- an overall score that summarizes the previous scores

```python
# add sentiment anaylsis columns
from nltk.sentiment.vader import SentimentIntensityAnalyzer as sti
import nltk
nltk.download('vader_lexicon')
sid = sti()

review_df["sentiments"] = review_df["Reviews"].apply(lambda x: sid.polarity_scores(x))
review_df = pd.concat([review_df.drop(['sentiments'], axis=1), review_df['sentiments'].apply(pd.Series)], axis=1)
```
